```python
import pandas as pd
```

```python
dataset=pd.read_csv("Data.csv",encoding="ISO-8859-1")
```

```python
data=dataset.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
new_columns=[str(i) for i in range(25)]
data.columns=new_columns
data.head()
```

```python
for i in new_columns:
    data[i]=data[i].str.lower()
data.head()
```

```python
headlines=[]
for i in range(len(data)):
    headlines.append(" ".join(str(x) for x in data.iloc[i,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset.columns
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv=CountVectorizer(ngram_range=(2,2))
```

```python
x=cv.fit_transform(headlines).toarray()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,dataset['Label'],test_size=0.2)
```

```python
model=RandomForestClassifier(n_estimators=200,criterion="entropy")
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_predicted)
```

```python
import seaborn as sn
```

```python
sn.heatmap(cm,annot=True)
```
